```python
import mysql.connector as myconn

mydb = myconn.connect(host="localhost", user="root", password="139400", database="ECOMM")

db_cursor = mydb.cursor()

try:
    db_cursor.execute("""
        CREATE TABLE customers (
            CustomerID INT PRIMARY KEY,
            Name VARCHAR(255),
            Email VARCHAR(255),
            JoinDate DATE
        )
    """)
    print("Table 'customers' created successfully.")
except myconn.Error as e:
    print("Error creating 'customers' table:", e)

try:
    db_cursor.execute("""
        CREATE TABLE products (
            ProductID INT PRIMARY KEY,
            Name VARCHAR(255),
            Category VARCHAR(255),
            Price DECIMAL(10, 2)
        )
    """)
    print("Table 'products' created successfully.")
except myconn.Error as e:
    print("Error creating 'products' table:", e)

try:
    db_cursor.execute("""
        CREATE TABLE orders (
            OrderID INT PRIMARY KEY,
            CustomerID INT,
            OrderDate DATE,
            TotalAmount DECIMAL(10, 2),
            FOREIGN KEY (CustomerID) REFERENCES customers(CustomerID)
        )
    """)
    print("Table 'orders' created successfully.")
except myconn.Error as e:
    print("Error creating 'orders' table:", e)

try:
    db_cursor.execute("""
        CREATE TABLE orderdetails (
            OrderDetailID INT PRIMARY KEY,
            OrderID INT,
            ProductID INT,
            Quantity INT,
            PricePerUnit DECIMAL(10, 2),
            FOREIGN KEY (OrderID) REFERENCES orders(OrderID),
            FOREIGN KEY (ProductID) REFERENCES products(ProductID)
        )
    """)
    print("Table 'orderdetails' created successfully.")
except myconn.Error as e:
    print("Error creating 'orderdetails' table:", e)
```

 # 2. Joins and Relationships

 #2.1. List all products along with the names of the customers who ordered them.

```python
import mysql.connector as myconn

mydb = myconn.connect(host="localhost", user="root", password="139400", database="ECOMM")

db_cursor = mydb.cursor()

db_cursor.execute("""
    SELECT p.ProductID, p.Name AS ProductName, c.Name AS CustomerName
    FROM Products p
    JOIN OrderDetails od ON p.ProductID = od.ProductID
    JOIN Orders o ON od.OrderID = o.OrderID
    JOIN Customers c ON o.CustomerID = c.CustomerID
""")

product_customers = db_cursor.fetchall()

for row in product_customers:
    print(row)
```

#2.2. Show orders that include more than one product.

```python
import mysql.connector as myconn

mydb = myconn.connect(host="localhost", user="root", password="139400", database="ECOMM")

db_cursor = mydb.cursor()

db_cursor.execute("""
    SELECT o.OrderID, COUNT(od.ProductID) AS NumberOfProducts
    FROM Orders o
    JOIN OrderDetails od ON o.OrderID = od.OrderID
    GROUP BY o.OrderID
    HAVING NumberOfProducts > 1
""")

multi_product_orders = db_cursor.fetchall()

for row in multi_product_orders:
    print(row)
```

 #3.1. Calculate the total revenue generated by each product category.

```python
import mysql.connector as myconn

mydb = myconn.connect(host="localhost", user="root", password="139400", database="ECOMM")

db_cursor = mydb.cursor()

db_cursor.execute("""
    SELECT p.Category, SUM(od.Quantity * od.PricePerUnit) AS TotalRevenue
    FROM Products p
    JOIN OrderDetails od ON p.ProductID = od.ProductID
    GROUP BY p.Category
""")

revenue_by_category = db_cursor.fetchall()

for row in revenue_by_category:
    print(row)
```

# 6.Advanced Queries:

#6.1. Rank customers based on their total spending

```python
import mysql.connector as myconn

mydb = myconn.connect(host="localhost", user="root", password="139400", database="ECOMM")

db_cursor = mydb.cursor()

db_cursor.execute("""
    SELECT c.CustomerID, c.Name AS CustomerName, SUM(o.TotalAmount) AS TotalSpending
    FROM Customers c
    JOIN Orders o ON c.CustomerID = o.CustomerID
    GROUP BY c.CustomerID
    ORDER BY TotalSpending DESC
""")

customer_ranking = db_cursor.fetchall()

for rank, customer in enumerate(customer_ranking, start=1):
    print("Rank:", rank)
    print("CustomerID:", customer[0])
    print("CustomerName:", customer[1])
    print("TotalSpending:", customer[2])
    print()
```

```python
#6.3. Calculate the month-over-month growth rate in sales.
```

```python
import mysql.connector as myconn

mydb = myconn.connect(host="localhost", user="root", password="139400", database="ECOMM")

db_cursor = mydb.cursor()

db_cursor.execute("""
    SELECT
        DATE_FORMAT(OrderDate, '%Y-%m') AS Month,
        SUM(TotalAmount) AS TotalSales
    FROM
        Orders
    GROUP BY
        DATE_FORMAT(OrderDate, '%Y-%m')
    ORDER BY
        DATE_FORMAT(OrderDate, '%Y-%m')
""")

monthly_sales = db_cursor.fetchall()

growth_rates = []
previous_sales = None
for month, sales in monthly_sales:
    if previous_sales is not None:
        growth_rate = ((sales - previous_sales) / previous_sales) * 100
        growth_rates.append((month, growth_rate))
    previous_sales = sales

for month, growth_rate in growth_rates:
    print("Month:", month)
    print("Growth Rate:", "{:.2f}%".format(growth_rate))
    print()
```
